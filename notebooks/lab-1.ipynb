## Task 3: Retrieve Sentinel-2 Data
Retrieve Sentinel-2 imagery for your defined AOI within a specific date range. Sort the images based on cloud cover percentage.

```python
# Retrieve Sentinel-2 Data
# TODO: Adjust the date range and retrieve the least cloudy Sentinel-2 image
sentinel2 = ee.ImageCollection('COPERNICUS/S2_HARMONIZED') .filterDate('2024-03-01', '2024-05-28') \
    .filterBounds(aoi)\
    .sort('CLOUDY_PIXEL_PERCENTAGE') 
#print(sentinel2.size().getInfo())

import geemap
image = sentinel2.first()
#image
metadata = image.getInfo()
metadata
```

## Task 4: Extract Image Metadata
Extract and print metadata from the selected image, such as ID, cloud cover percentage, mean solar azimuth and zenith angles, and processing level.

```python
# Extract metadata
# TODO: Retrieve and display image metadata
product_id = metadata['properties'].get('PRODUCT_ID')
product_id
cloud_coverage = metadata['properties']['CLOUDY_PIXEL_PERCENTAGE']
cloud_coverage
MEAN_SOLAR_AZIMUTH_ANGLE = metadata['properties']['MEAN_SOLAR_AZIMUTH_ANGLE']
MEAN_SOLAR_ZENITH_ANGLE = metadata['properties']['MEAN_SOLAR_ZENITH_ANGLE']
PROCESSING_BASELINE = metadata['properties']['PROCESSING_BASELINE']
print(f"Product ID: {product_id}")
print(f"Cloud coverage: {cloud_coverage}%") 
print(f"Solar azimuth angle: {MEAN_SOLAR_AZIMUTH_ANGLE}")
print(f"Solar zenith angle: {MEAN_SOLAR_ZENITH_ANGLE}")
print(f"Processing baseline: {PROCESSING_BASELINE}")
```
