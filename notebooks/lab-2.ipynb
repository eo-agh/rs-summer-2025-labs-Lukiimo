## Task 3: Calculate NDVI
Calculate the Normalized Difference Vegetation Index (NDVI) for both images.

```python
# TODO: Calculate NDVI for both images (Choose the appropriate S-2 bands).
image = image.clip(aoi)
image1 = image1.clip(aoi)


ndvi = image.normalizedDifference(['B8', 'B4']).rename('NDVI')
ndvi2 = image1.normalizedDifference(['B8', 'B4']).rename('NDVI')

# Add NDVI layer to the map
Map.addLayer(ndvi, {'min': 0, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'NDVI')
Map.addLayer(ndvi2, {'min': 0, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'NDVI2')
Map
```

## Task 4: Compare NDVI
Display both NDVI indices on the map and calculate and visualize their difference.

```python
# Assuming ndvi1 and ndvi2 are ee.Image objects
ndvi_diff = ndvi2.subtract(ndvi).rename('NDVI_Diff')

Map.addLayer(ndvi_diff, {'min': 0, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'ndvi_diff')
Map
```
